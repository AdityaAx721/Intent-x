```python
import json
from datasets import load_dataset

dataset = load_dataset("banking77")

labels = dataset["train"].features["label"].names

intent_map = {str(i): label for i, label in enumerate(labels)}

with open("models/intent_mapping.json", "w") as f:
    json.dump(intent_map, f, indent=2)

print("intent_mapping.json created successfully")
```
